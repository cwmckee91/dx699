```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

# How to get the user to examine the last 3 months first

1. User color contrast - ie highlight the last three months in a bolder color while keeping earlier months in a separate more dull color

2. Increase the line thickness or point size of the last three months, changing color would also be ideal in a similar fashion to point #1

3. Annotate the key points

4. Change the background of the grid to emphasize those months

```python
np.random.seed(1)
numdates = 12
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')

ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")

# Color contrast: earlier months in light gray, last three months in red
ax.plot(datearray[:-3], time_series[:-3], color="lightgray")
ax.plot(datearray[-3:], time_series[-3:], color="red", linewidth=2)
```

```python
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")

# Thicker lines for the last three months
ax.plot(datearray[:-3], time_series[:-3], color="blue", linewidth=1)
ax.plot(datearray[-3:], time_series[-3:], color="red", linewidth=5)
```

```python
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")

# Annotate the last three months
ax.annotate("Recent peak", xy=(datearray[-1], time_series[-1]), xytext=(-20, 10),
            textcoords="offset points", arrowprops=dict(arrowstyle="->", color="red"))
```

```python
# Page 114

import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
received = [150, 180, 250, 160, 190, 140, 200, 170, 150, 140, 160, 180]
processed = [140, 170, 240, 150, 180, 130, 190, 160, 140, 130, 150, 170]

plt.figure(figsize=(10, 5))
plt.plot(months, received, label="Received", color="gray", linewidth=2)
plt.plot(months, processed, label="Processed", color="navy", linewidth=2)

plt.xlabel("")
plt.ylabel("")
plt.xticks(rotation=0)
plt.legend(loc="best", fontsize=12)
plt.ylim(0, 300)
plt.grid(False)
plt.show()
```
